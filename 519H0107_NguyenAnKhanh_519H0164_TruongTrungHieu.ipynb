```python
def get_train_data(days, data1):
    data_len = data1.shape[0]
    X_train = []
    y_train = []
    for i in range(days, data_len):
        X_train.append(data1[i-days:i])
        y_train.append(data1[i])
    return np.array(X_train), np.array(y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import SimpleRNN
from sklearn.neural_network import MLPClassifier
from tensorflow.keras.layers import Dropout
import keras
from keras.callbacks import EarlyStopping
```

```python
def RNN_with_Dropout(X_train, y_train, epochs, batch_size, dropout_rate):
        rnn = Sequential()
        rnn.add(SimpleRNN(units=50, activation='relu',
                input_shape=(n_input, n_features), return_sequences=True))
        rnn.add(Dropout(dropout_rate))

        rnn.add(SimpleRNN(units=50, return_sequences=True))
        rnn.add(Dropout(dropout_rate))

        rnn.add(SimpleRNN(units=50, return_sequences=False))
        rnn.add(Dropout(dropout_rate))
        
        rnn.add(Dense(4))

        callback = EarlyStopping(monitor='loss', patience=15)

        rnn.compile(optimizer='adam', loss='mse',
                    metrics=['mean_absolute_error', 'accuracy'])
        rnn.fit(X_train, y_train,
                epochs=epochs, batch_size=batch_size, callbacks = [callback])
        return rnn
```

```python
def RNN(X_train, y_train, epochs, batch_size):
    rnn = Sequential()
    rnn.add(SimpleRNN(units=50, activation='relu',
                      input_shape=(n_input, n_features), return_sequences=True))

    rnn.add(SimpleRNN(units=50, return_sequences=True))
#     rnn.add(SimpleRNN(units=50, return_sequences=True))
#     rnn.add(SimpleRNN(units=50, return_sequences=True))
#     rnn.add(SimpleRNN(units=50, return_sequences=True))
    rnn.add(SimpleRNN(units=50, return_sequences=True))

    rnn.add(SimpleRNN(units=50, return_sequences=False))
    rnn.add(Dense(4))

    callback = EarlyStopping(monitor='loss', patience=15)

    rnn.compile(optimizer='adam', loss='mse', metrics=['mean_absolute_error', 'accuracy'])
    rnn.fit(X_train, y_train,
            epochs=epochs, batch_size=batch_size, callbacks=[callback])
    return rnn
```

```python
from keras.optimizers import RMSprop
def MLP_with_Dropout(X_train, y_train, epochs, batch_size, dropout_rate):
    mlp = Sequential()
    mlp.add(Flatten(input_shape=(n_input, n_features)))

    mlp.add(Dense(64, activation='relu'))
    mlp.add(Dropout(dropout_rate))
    mlp.add(Dense(64, activation='relu'))
    mlp.add(Dropout(dropout_rate))
    mlp.add(Dense(64, activation='relu'))
    mlp.add(Dropout(dropout_rate))
    mlp.add(Dense(64, activation='relu'))
    mlp.add(Dropout(dropout_rate))
    mlp.add(Dense(128, activation='relu'))
    mlp.add(Dropout(dropout_rate))
    mlp.add(Dense(128, activation='relu'))
    mlp.add(Dropout(dropout_rate))
    mlp.add(Dense(128, activation='relu'))
    mlp.add(Dropout(dropout_rate))
    mlp.add(Dense(128, activation='relu'))
    mlp.add(Dropout(dropout_rate))
    mlp.add(Dense(256, activation='relu'))
    mlp.add(Dropout(dropout_rate))

    mlp.add(Dense(4, activation='softmax'))

    callback = EarlyStopping(monitor='loss', patience=15)

    mlp.compile(
        loss='mse',
        optimizer=RMSprop(),
        metrics=['mean_absolute_error', 'accuracy']
    )
    mlp.fit(
        X_train, y_train,
        batch_size=batch_size,
        epochs=epochs,
        verbose=1,
        callbacks=[callback]
    )

    return mlp
```

```python
from keras.optimizers import RMSprop


def MLP(X_train, y_train, epochs, batch_size):
    mlp = Sequential()
    mlp.add(Flatten(input_shape=(n_input, n_features)))
    mlp.add(Dense(64, activation='relu'))
    mlp.add(Dense(64, activation='relu'))
    mlp.add(Dense(64, activation='relu'))
    mlp.add(Dense(64, activation='relu'))
    mlp.add(Dense(128, activation='relu'))
    mlp.add(Dense(128, activation='relu'))
    mlp.add(Dense(128, activation='relu'))
    mlp.add(Dense(128, activation='relu'))
    mlp.add(Dense(256, activation='relu'))
    mlp.add(Dense(4, activation='softmax'))

    callback = EarlyStopping(monitor='loss', patience=15)

    mlp.compile(
        loss='mse',
        optimizer=RMSprop(),
        metrics=['mean_absolute_error', 'accuracy']
    )
    mlp.fit(
        X_train, y_train,
        batch_size=batch_size,
        epochs=epochs,
        verbose=1,
        callbacks=[callback]
    )
    return mlp
```

```python
def predict_1day(model, test_data, company):
    test_60_days = test_data[-60:]

    if (company == 'HPG'):
        test_60_days = sc_HPG.transform(test_60_days)
    elif (company == 'FPT'):
       test_60_days = sc_FPT.transform(test_60_days)
    elif (company == 'STB'):
        test_60_days = sc_STB.transform(test_60_days)
    else:
        test_60_days = sc_VIC.transform(test_60_days)

    X_test2 = []
    X_test2.append(test_60_days)
    X_test2 = np.array(X_test2)
    pred_price = model.predict(X_test2)
    if(company == 'HPG'):
        pred_price = sc_HPG.inverse_transform(pred_price)
    elif(company == 'FPT'):
        pred_price = sc_FPT.inverse_transform(pred_price)
    elif(company == 'STB'):
        pred_price = sc_STB.inverse_transform(pred_price)
    else:
        pred_price = sc_VIC.inverse_transform(pred_price)
    pred_price = pd.DataFrame(pred_price)
    pred_price.rename(columns={0: 'Open', 1: 'Highest',
                               2: 'Lowest', 3: 'Close'}, inplace=True)
    return pred_price
```
